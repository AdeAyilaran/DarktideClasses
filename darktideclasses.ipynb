```python
# Create the ability to colour the terminal using ANSI Escape Codes

class colours:
    GREEN = '\u001b[32m'
    RED = '\u001b[31m'
    BLUE = '\u001b[34m'
    BLACK = '\u001b[30m'
```

```python
# Modules to import

import random
import inspect
```

```python
# Make the parent class

class Character:
    
    def __init__(self, name, health = 100.0, maxhealth = 100.0, toughness = 100.0,
                maxtoughness = 100.0, damage = 10.0, ml = 1, mu = 1, rl = 1, ru = 1, ap = 3):
        
        self.name = name
        self.health = health
        self.maxhealth = maxhealth
        self.toughness = toughness
        self.maxtoughness = maxtoughness
        self.damage = damage
        self.ml = ml
        self.mu = mu
        self.rl = rl
        self.ru = ru
        self.ap = ap
        
    def melee_attack(self, target):
        
        if self.ap > 0:
            melee_damage = self.damage*random.randint(self.ml, self.mu)
            if target.toughness > 0.0:
                get_toughness = (target.toughness - melee_damage)
                if get_toughness > 0.0:
                    target.health -= (1.0 - 0.9*(target.toughness/target.maxtoughness))*melee_damage
                    target.toughness -= melee_damage
                    print(colours.RED + f"{self.name} just attacked {target.name} for {melee_damage} points of damage!")
                    self.ap -= 1
                else:
                    get_health = (melee_damage - target.toughness)
                    target.toughness = 0.0
                    target.health -= get_health
                    print(colours.RED + f"{self.name} just attacked {target.name} for {melee_damage} points of damage!")
                    if target.health < 0.0:
                        target.health = 0.0
                    self.ap -= 1
            else:
                target.health -= melee_damage
                print(colours.RED + f"{self.name} just attacked {target.name} for {melee_damage} points of damage!")
                if target.health < 0.0:
                    target.health = 0.0
                self.ap -= 1   
        else:
            print(colours.BLUE + f"{self.name} has no Action Points!!")
        
        
    def ranged_attack(self, target):
        
        if self.ap > 0:
            range_damage = self.damage*random.randint(self.rl, self.ru)
            if target.toughness > 0.0:
                get_toughness = (target.toughness - range_damage)
                if get_toughness > 0.0:
                    target.toughness -= range_damage
                    print(colours.RED + f"{self.name} just attacked {target.name} for {range_damage} points of damage!")
                    self.ap -= 1
                else:
                    get_health = (range_damage - target.toughness)
                    target.toughness = 0.0
                    target.health -= get_health
                    print(colours.RED + f"{self.name} just attacked {target.name} for {range_damage} points of damage!")
                    if target.health < 0.0:
                        target.health = 0.0
                    self.ap -= 1
            else:
                target.health -= range_damage
                print(colours.RED + f"{self.name} just attacked {target.name} for {range_damage} points of damage!")
                if target.health < 0.0:
                    target.health = 0.0
                self.ap -= 1  
        else:
            print(colours.BLUE + f"{self.name} has no Action Points!!")
            
    def reset(self, target):
        if self.ap > 0:
            self.health += 20.0
            if self.health > self.maxhealth:
                self.health = self.maxhealth
            self.ap -=1
        else:
            print(colours.BLUE + f"{self.name} has no Action Points!!")
            
    def regen_toughness(self):
        self.toughness += 0.2*self.maxtoughness
        if self.toughness > self.maxtoughness:
            self.toughness = self.maxtoughness
        
    def reset_ap(self):
        print(colours.BLUE + "Refreshing Action Points!")
        self.ap = 3
        
    def display_stats(self):
        print(colours.GREEN + f"Current {self.name} Stats: \n Health: {self.health} \n Toughness: {self.toughness} \n Action Points: {self.ap}")
        
    def choose_action(self, target):
        
        all_methods = inspect.getmembers(self, predicate = inspect.ismethod)
        
        method = [method[0] for method in all_methods if method[0] != '__init__' and method[0] != 'display_stats'
                  and method[0] != 'choose_action' and method[0] != 'reset_ap' and method[0] != 'regen_toughness']
        
        random_action = random.choice(method)
        print(colours.GREEN + f"{self.name} chose {random_action}!")
        getattr(self, random_action)(target)
```

```python
# Making the child class Sharpshooter

class Sharpshooter(Character):
    
    def __init__(self, name, health = 125.0, maxhealth = 125.0, toughness = 200.0,
                maxtoughness = 200.0, damage = 10.0, ml = 2, mu = 3, rl = 2, ru = 4, ap = 3):
        super().__init__(name, health, maxhealth, toughness, maxtoughness, damage, ml, mu, rl, ru, ap)
        self.name = name
        self.health = health
        self.maxhealth = maxhealth
        self.toughness = toughness
        self.maxtoughness = maxtoughness
        self.damage = damage
        self.ml = ml
        self.mu = mu
        self.rl = rl
        self.ru = ru
        self.ap = ap
        
    def volley_fire(self, target):
        if self.ap == 3:
            x = random.randint(2, 5)
            for i in range(x):
                self.ap += 1
                print(colours.RED + "Volley Fire!")
                self.ranged_attack(target)
            self.ap = 1
        elif self.ap == 2:
            x = random.randint(2, 5)
            for i in range(x):
                self.ap += 1
                print(colours.RED + "Volley Fire!")
                self.ranged_attack(target)
            self.ap = 0
        else:
            print(colours.BLUE + f"{self.name} needs at least 2 Action Points!")
            
    
```

```python
# Making the child class Preacher

class Preacher(Character):
    
    def __init__(self, name, health = 200.0, maxhealth = 200.0, toughness = 100.0,
                maxtoughness = 100.0, damage = 17.0, ml = 2, mu = 6, rl = 2, ru = 3, ap = 3):
        super().__init__(name, health, maxhealth, toughness, maxtoughness, damage, ml, mu, rl, ru, ap)
        self.name = name
        self.health = health
        self.maxhealth = maxhealth
        self.toughness = toughness
        self.maxtoughness = maxtoughness
        self.damage = damage
        self.ml = ml
        self.mu = mu
        self.rl = rl
        self.ru = ru
        self.ap = ap
        
    def chastise_the_wicked(self, target):
        if self.ap > 0:
            self.toughness += 0.5*self.maxtoughness
            if self.toughness > self.maxtoughness:
                self.toughness = self.maxtoughness
            print(colours.GREEN + f"{self.name} regenerated {0.5*self.maxtoughness} of toughness!")
            self.melee_attack(target)
        else:
            print(colours.BLUE + f"{self.name} has no Action Points!")
```

```python
class PlagueOgryn(Character):
    
    def __init__(self, name, health = 500.0, maxhealth = 500.0, toughness = 0.0,
                maxtoughness = 0.0, damage = 35.0, ml = 2, mu = 6, rl = 1, ru = 3, ap = 3):
        super().__init__(name, health, maxhealth, toughness, maxtoughness, damage, ml, mu, rl, ru, ap)
        self.name = name
        self.health = health
        self.maxhealth = maxhealth
        self.toughness = toughness
        self.maxtoughness = maxtoughness
        self.damage = damage
        self.ml = ml
        self.mu = mu
        self.rl = rl
        self.ru = ru
        self.ap = ap
        
    def disease_cloud(self, target):
        print(colours.RED + f"{self.name} used Disease Cloud on {target.name}!")
        target.health -= 0.2*target.maxhealth
```

```python
rogue = Sharpshooter("Rouge")
scourge = Preacher("Scourge")
plagueogryn = PlagueOgryn("Plague Ogryn")
dreg = Character(name = "Dreg Gunner", health = 200.0, maxhealth = 200.0, toughness = 0.0, maxtoughness = 0.0,
                 damage = 16.0, ml = 2, mu = 4, rl = 2, ru = 4)

ally = [rogue, scourge]
enemy = [plagueogryn, dreg]
```

```python
while True:
    
    print(f"---- Round {round} -----")
    
    health_zero = [character.health for character in (ally + enemy)]
    toughness_zero = [character.toughness for character in (ally + enemy)]
    print(colours.BLACK + f"All Health: {health_zero}")
    print(colours.BLACK + f"All Toughness: {toughness_zero}")
    if 0.0 in health_zero:
        print("Someone has died!")
        print("Game Over!")
        break
        
    for character in ally:
        character.choose_action(random.choice(enemy))
        character.display_stats()
        
    health_zero = [character.health for character in (ally + enemy)]
    toughness_zero = [character.toughness for character in (ally + enemy)]
    print(colours.BLACK + f"All Health: {health_zero}")
    print(colours.BLACK + f"All Toughness: {toughness_zero}")
    if 0.0 in health_zero:
        print("Someone has died!")
        print("Game Over!")
        break
        
    for character in enemy:
        character.choose_action(random.choice(ally))
        character.display_stats()
        
    health_zero = [character.health for character in (ally + enemy)]
    toughness_zero = [character.toughness for character in (ally + enemy)]
    print(colours.BLACK + f"All Health: {health_zero}")
    print(colours.BLACK + f"All Toughness: {toughness_zero}")
    if 0.0 in health_zero:
        print("Someone has died!")
        print("Game Over!")
        break
        
    ap_zero = all(character.ap == 0 for character in (ally + enemy))
    
    if ap_zero == True:
        for character in (ally + enemy):
            character.reset_ap()
            
        round += 1
        print("\n--- Next Round ---\n")
    
```
